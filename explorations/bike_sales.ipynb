```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
df = pd.read_csv('../data/Sales.csv')

df.head()
```

# Bar Chart Of The Diffrent Age Groups

```python

total_age_adults =  len(df[df["Age_Group"] == "Adults (35-64)"])
total_age_youth =  len(df[df["Age_Group"] == "Youth (<25)"])

x = ["Age (35-64)", "Age (0-25)"]
y = total_age_adults, total_age_youth


plt.bar(x,y)
plt.show()
```
